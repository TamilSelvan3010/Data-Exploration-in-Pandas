## About The Dataset

##### The Atlantic and Pacific cod landings NOAA dataset is a collection of data on the fishing industry for Atlantic and Pacific cod from 1950 to 2021. It includes information on the year, state, and scientific name of the fish, as well as the pounds and metric tons of the catch and the dollars earned from the sale of the fish. The data in this dataset can be used to track the trends in cod fishing over time and to understand the economic impact of this industry in different states and regions. It can also be used to inform conservation and management efforts for these species.
